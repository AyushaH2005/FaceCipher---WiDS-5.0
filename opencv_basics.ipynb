```python
img = cv2.imread("img/img1.jpg")
#cv2.imshow("window", img)
#cv2.waitKey(0)
img_flip_1 = cv2.flip(img,-1)
cv2.imshow("window", img_flip_1)
cv2.waitKey(0)
```

```python
img_slice = img[0:1000,0:1500]
cv2.imshow("window", img_slice)
cv2.waitKey(0)
```

```python
height, width,_ = img.shape
T = cv2.getRotationMatrix2D((width/2,height/2),90,1)
img_trans = cv2.warpAffine(img,T,(width,height))
cv2.imshow("window", img_trans)
cv2.waitKey(0)
```

```python
img_own = np.zeros((512,512,3))
# Rectangle
cv2.rectangle(img_own,pt1=(100,100),pt2=(300,200),color=(255,0,0), thickness = 1)
# Circle
cv2.circle(img_own,center = (60,60),radius = 50,color=(0,0,255),thickness = 1)
#Line
cv2.line(img_own, pt1 = (0,0), pt2 = (512,512),color = (0,255,0), thickness = 1)
#Text
cv2.putText(img_own, org=(300,300), fontScale = 1, color=(255,255,0), thickness = 1, text="Hello World!", lineType = cv2.LINE_AA,fontFace = cv2.FONT_ITALIC)
cv2.imshow("window",img_own)
cv2.waitKey(0)
```

```python
k_size = 9
img = cv2.imread("img/img2.jpg")
img_blur = cv2.blur(img,(k_size,k_size))
img_gaussian = cv2.GaussianBlur(img,(k_size,k_size),5)
img_median = cv2.medianBlur(img,k_size)
cv2.imshow("window",img_median)
cv2.waitKey(0)
```

```python
#Simple thresholding (Binary)
img = cv2.imread("img/img1.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(img_gray, 80, 255, cv2.THRESH_BINARY)
thresh = cv2.blur(thresh, (10,10))
ret, thresh = cv2.threshold(thresh, 80, 255, cv2.THRESH_BINARY)
print(ret)
cv2.imshow("window", thresh)
cv2.waitKey(0)
```

```python
#Adaptive threshold
img = cv2.imread("img/img3.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
thresh = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 25, 29)
cv2.imshow("window", thresh)
cv2.waitKey(0)
```

```python
img = cv2.imread("img/img2.jpg")
img_edge = cv2.Canny(img, 100, 200)
cv2.imshow("window", img_edge)
cv2.waitKey(0)
```

```python
img = cv2.imread("img/img4.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(img_gray,127,255, cv2.THRESH_BINARY_INV)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
for cnt in contours:
    if cv2.contourArea(cnt) >50:
        cv2.drawContours(img, cnt, -1, (0, 255, 0), 1)
    print(cv2.contourArea(cnt))
cv2.imshow("img", img)
#cv2.imshow("thresh", contours)
cv2.waitKey(0)
```
